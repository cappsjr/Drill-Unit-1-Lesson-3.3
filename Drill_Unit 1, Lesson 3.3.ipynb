```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/Users/cappsjr/Downloads/happiness_report.csv')
```

```python
sns.distplot(df['Happiness Score'], bins = 8, kde = False, rug = True)
```

This histogram shows the distribution of the data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.  The number of bins can be set, but the default behavior is to guess for the right number.  This particular histogram also has a rug plot, which places a small tick mark at the bottom of the graph at each observation.

```python
sns.kdeplot(df['Happiness Score'], shade=True)
```

The kernel density distribution is good for plotting the shape of a distribution.  A bandwidth parameter can be used to control how tightly the estimation is fit to the data.  Like bins for a histogram, bandwidth can be set, but the default is to guess a good value.  

```python
sns.distplot(df['Happiness Score'], rug = True)
```

Using the distplot function, you can fit a parametric distribtuion to see how closely it fits to the observed data.  

```python
sns.boxplot(df['Happiness Score'])
```

The boxplot function shows the distribution of the data.  The left and right edges of the box mark the interquartile range, while the line in the middle of the box marks the mean of the data.  The horizontal line that connects the box to the outer vertical lines denotes the range of the data from the interquartile range to the most extreme high and low data points.  

```python
sns.jointplot(x = df['Happiness Score'], y = df['Economy (GDP per Capita)'])
```

The scatter plot is like a rug plot on both the x-axis and the y-axis where each data point is shown as an x-value and a y-value.  This is a great choice to show the relationship between two continuous variables variables.

```python
sns.jointplot(x = df['Happiness Score'], y = df['Economy (GDP per Capita)'], kind = 'hex')
```

This bivariate version of the histogram is called a hexbin.  It shows the number of observations that fall into hexagonal bins.  The darker the hexagon the more observations that fall in that bin.  

```python
sns.jointplot(x = df['Happiness Score'], y = df['Economy (GDP per Capita)'], kind = 'kde')
```

This visualization uses the kernel density estimation to show a bivariate distribution.  It shows a summary of the distribution using a contour plot.  

```python
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips)
```

This strip plot shows the relationship between the categorical variable "day" and the continuous variable "total bill."  In a strip plot, the data points can overlap, making it difficult to see the full distribution of the data.

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

This is another strip plot showing the relationship between "day" and "total bill," but in order to more see the distribution of the data more fully, I added jitter, which adjusts the data along the categorical axis.

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

Like the strip plot, the swarm plot positions the data along the categorical axis, but it also avoids overlap.

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

It is also possible to added a nested categorical variable to a swarm plot, in this case male and female, using the hue parameter.  

```python
sns.violinplot(x="total_bill", y="day", hue="time", data=tips)
```

The violin plot incorporates aspects of the boxplot and the swarm plot by combining a boxplot with the kernel density estimation used in distributions.  The kernel density estimation provides a more accurate distribution of the data, and the quartile and whisker data are found inside the violin.

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="w", alpha=.5)
```

By combining the violin plot with the swarm plot, this visualization allows the observer to see the individual data points while still getting a summary of the distribution.
